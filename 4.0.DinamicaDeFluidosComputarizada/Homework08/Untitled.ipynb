```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sci
plt.rc('xtick', labelsize=18)     
plt.rc('ytick', labelsize=18)

df = pd.read_csv("./export.csv", delimiter=",");
print(df.columns)

plt.figure()
scale = 6;
plt.subplots(figsize=(3*scale, 2*scale));
ax0 = plt.gca()

# Calcualte the time required to cool down to reach freezing point at the surface.
temp = df.iloc[(df['surfaceTemperature_sphere']-0).abs().argsort()[:1]]
plt.scatter(temp["time_sphere"], temp["surfaceTemperature_sphere"], label="surface freezing point T" + r'$=0^\circ C$' + " at t=" + r'$%.2fs$' % (temp["time_sphere"]))

# Estimate the heat rate, and temperature as a function in time, for the center and surface.

plt.plot(df["time_sphere"], df["centerTemperature_sphere"], label="sphere center temperature")
plt.plot(df["time_sphere"], df["surfaceTemperature_sphere"], label="sphere surface temperature")

# Estimate the average temeprature as a function of time of the fruit.
plt.plot(df["time_sphere"], df["avgTemperature_sphere"], label="sphere average temperature")

# Estimate the total heat lost.


plt.legend(prop={'size': 16}, frameon=False)

ax0.tick_params(
        which='both',
        direction='in',
        # length=5,
        width=2,
        bottom=True,
        top=True,
        left=True,
        right=True);

plt.tight_layout()
plt.xlabel("time" + '    ' + r'$[s]$'              , fontsize=24);
plt.ylabel("temperature" + '    ' + r'$[^\circ C]$', fontsize=24);

plt.savefig('SphereTemperatures.png', dpi=200, bbox_inches='tight')  
```
