```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sci
plt.rc('xtick', labelsize=24)     
plt.rc('ytick', labelsize=24)
```

```python
def lucasWashburn(t):
    gamma = 0.072 # surface tension
    r     = 0.00025 # capillary radius
    theta = 100 # meniscus angle
    eta   = 0.001003*2500 # viscosity
    z     = 0.006 # height
    rho   = 998.2 # density
    g     = 9.81
    
    nume1 = t * gamma * r * np.cos(theta)
    deno1 = 4 * eta * z
    nume2 = t * r**2 * rho * g
    deno2 = 8 * eta
    
    return np.sqrt( (nume1 / deno1) - (nume2 / deno2) )# / 50
```

```python
plt.figure()
scale = 6;
plt.subplots(figsize=(3*scale, 2*scale));
ax0 = plt.gca()

t = np.linspace(0, 0.15, 1000);
h = lucasWashburn(t)
plt.plot(t, h, label="Lucas-Washburn equation")
plt.scatter(
    [0.0, 0.03, 0.06, 0.09, 0.13, 0.16],
    [0.0000, 0.0014, 0.0028, 0.0040, 0.0048, 0.0058],
    label="CFD results"
)

plt.legend(prop={'size': 24}, frameon=False)

ax0.tick_params(
        which='both',
        direction='in',
        # length=5,
        width=2,
        bottom=True,
        top=True,
        left=True,
        right=True);

plt.tight_layout()
plt.xlabel("time" + '    ' + r'$[s]$'              , fontsize=24);
plt.ylabel("height" + '    ' + r'$[m]$', fontsize=24);

plt.savefig('modelVScfd.png', dpi=200, bbox_inches='tight')  
```
