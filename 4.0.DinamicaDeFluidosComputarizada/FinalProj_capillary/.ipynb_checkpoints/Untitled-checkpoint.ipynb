```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sci
plt.rc('xtick', labelsize=18)     
plt.rc('ytick', labelsize=18)
```

```python
def lucasWashburn(t):
    gamma = 0.072 # surface tension
    r     = 0.00025 # capillary radius
    theta = 20 # meniscus angle
    eta   = 0.001003 # viscosity
    z     = 0.006 # height
    rho   = 998.2 # density
    g     = 9.81
    
    nume1 = t * gamma * r * np.cos(theta)
    deno1 = 4 * eta * z
    nume2 = t * r**2 * rho * g
    deno2 = 8 * eta
    
    return np.sqrt( (nume1 / deno1) - (nume2 / deno2) )
```

```python
plt.figure()
scale = 6;
plt.subplots(figsize=(3*scale, 2*scale));
ax0 = plt.gca()

t = np.linspace(0, 0.1, 1000);
h = lucasWashburn(t)
plt.plot(t, h, label="sphere average temperature")

plt.legend(prop={'size': 16}, frameon=False)

ax0.tick_params(
        which='both',
        direction='in',
        # length=5,
        width=2,
        bottom=True,
        top=True,
        left=True,
        right=True);

plt.tight_layout()
plt.xlabel("time" + '    ' + r'$[s]$'              , fontsize=24);
plt.ylabel("temperature" + '    ' + r'$[^\circ C]$', fontsize=24);

plt.savefig('SphereTemperatures.png', dpi=200, bbox_inches='tight')  
```
