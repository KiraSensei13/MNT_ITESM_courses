```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from statsmodels.regression.linear_model import OLS
from statsmodels.tools.tools import add_constant
```

```python
kidney_data = pd.read_csv('kidney.csv')
kidney_data.plot(x='age', y='tot', kind='scatter', figsize=(20,10))
```

```python
# Intercept column
kidney_data['intercept'] = 1
# Compute columns for age_squared and age_cubed
kidney_data['age_sq'] = kidney_data['age'] ** 2
kidney_data['age_cu'] = kidney_data['age'] ** 3

# Since the three predictor variables are on quite different scales, we'll need to center and scale them...
age_mean, age_std = kidney_data['age'].mean(), kidney_data['age'].std()
age_sq_mean, age_sq_std = kidney_data['age_sq'].mean(), kidney_data['age_sq'].std()
age_cu_mean, age_cu_std = kidney_data['age_cu'].mean(), kidney_data['age_cu'].std()

# Not subtracting mean age from age variable, since I want that to be captured by the intercept
kidney_data['age'] = (kidney_data['age']) / age_std
kidney_data['age_sq'] = (kidney_data['age_sq'] - age_sq_mean) / age_sq_mean
kidney_data['age_cu'] = (kidney_data['age_cu'] - age_cu_mean) / age_cu_std
```

```python
# Fit an Ordinary Least Squares (OLS) Multiple Regression estimator
regressor_columns = ['intercept', 'age', 'age_sq', 'age_cu']
ols_reg_model = OLS(kidney_data['tot'], kidney_data[regressor_columns]).fit()
ols_reg_model.summary()
```

```python
# None of the regression coefficients are statistically significant now, but let's keep them just to show how to get confidence intervals for predictions on multiple dimensions.
kidney_data['tot_hat_cureg'] = ols_reg_model.predict()
plt = matplotlib.pyplot
plt.figure(figsize=(20,10))
plt.scatter(kidney_data['age'], kidney_data['tot'],
         marker='o', color='skyblue', linewidth=2)
plt.plot(kidney_data['age'], kidney_data['tot_hat_cureg'], color='olive')
plt.legend()
plt.show()
```

```python
# Estimate error variance sigma^2 using (sum of squared errors) / (n - p - 1), where p=3, since we estimated 3 betas.
residuals = ols_reg_model.predict() - kidney_data['tot']
n = kidney_data.shape[0]
p = len(regressor_columns) - 1
sigma_sq_hat = (residuals ** 2).sum() / (n - p - 1)
```

```python
# Next, calculate (X'X)^(-1), where X is our matrix of predictors age, age_sq, age_cu
X_mat = np.array(kidney_data[regressor_columns])
X_sq_mat_inv = np.linalg.inv(np.matmul(np.transpose(X_mat), X_mat))
```

```python
selected_age_values = np.reshape(np.linspace(20,80,7), (7,1))
test_X_mat = np.concatenate((np.reshape(np.repeat(1,(7)), (7,1)), selected_age_values), axis=1)
test_X_mat[:,0] = 1
test_X_mat[:,1] = (test_X_mat[:,1]) / age_std
if 'age_sq' in regressor_columns:
    test_X_mat = np.concatenate((test_X_mat, selected_age_values ** 2), axis=1)
    test_X_mat[:,2] = (test_X_mat[:,2] - age_sq_mean) / age_sq_std
if 'age_cu' in regressor_columns:
    test_X_mat = np.concatenate((test_X_mat, selected_age_values ** 3), axis=1)
    test_X_mat[:,3] = (test_X_mat[:,3] - age_cu_mean) / age_cu_std
test_X_mat
```

```python
# Finally, let's get the y_hat standard errors for the requested values of 20, 30, 40, ..., 80.
n_test = selected_age_values.shape[0]
se_y_hats = np.zeros((n_test,1))
for i in range(n_test):
    se_y_hats[i] = np.sqrt(sigma_sq_hat * np.matmul(np.matmul(np.reshape(test_X_mat[i],(1,len(regressor_columns))), X_sq_mat_inv), np.reshape(test_X_mat[i], (len(regressor_columns),1)))[0][0])
```

```python
# In general, the SE's look quite higher.
se_y_hats
```
