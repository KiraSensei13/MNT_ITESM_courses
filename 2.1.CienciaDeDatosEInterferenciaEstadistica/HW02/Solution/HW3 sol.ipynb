```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

### Part a) Run simulation 200 times

```python
ind = np.linspace(1,500,500)
mu = ind * 3 / 500
d_list = list()
for rep in range(1,200):
    x_vec = norm.rvs(loc=mu, scale=1)
    i_max = np.argmax(x_vec)
    d = x_vec[i_max] - mu[i_max]
    d_list.append(d)
```

### Part b) Plot histogram of d

```python
pd.DataFrame({'d_values': d_list}).plot(kind='hist', figsize=(20,10))
```

### Part c): It is showing that if we pick the value of an effect from a sample because it was the largest observed effect, that is highly likely to overestimate the true value of the effect.
