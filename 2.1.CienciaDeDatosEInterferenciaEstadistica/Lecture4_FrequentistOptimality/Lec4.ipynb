```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# Define values of c to be plugged in, as well as empty arrays for alphas and betas
n = 100
c_interv = [.8,.7,.6,.5,.4,.3,.25,.2,.1,0,-.1,-.2,-.3,-.4,-.5,-.6,-.7,-.8]
alphas = np.array([])
betas = np.array([])
```

```python
# Calculate and save alpha and beta for each c
```

```python
for c in c_interv:
    alpha = 1 - norm.cdf(c * np.sqrt(n))
    beta = norm.cdf((c - 0.5)* np.sqrt(n))
    alphas = np.append(alphas, alpha)
    betas = np.append(betas, beta)
```

```python
# Now plot alphas, betas and c's
plt.figure(figsize=(15,15))
plt.scatter(x=alphas, y=betas, color='red', s=50)
plt.plot(alphas, betas, color='black')
plt.xticks(np.linspace(0,1,11))
plt.yticks(np.linspace(0,1,11))
plt.xlabel(r'$\alpha$', fontsize=24)
plt.ylabel(r'$\beta$', fontsize=24)
plt.grid()
for ind, c in enumerate(c_interv):
    plt.annotate('    c = ' + str(np.round(c,2)), (alphas[ind], betas[ind]), fontsize=15)

plt.show()
```
