```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.special import gammaln
```

```python
# Beta Distributions
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.plot(np.linspace(0.01,0.99,200), st.beta.pdf(np.linspace(0.01,0.99,200), 1/2,1/2))
ax.plot(np.linspace(0.01,0.99,200), st.beta.pdf(np.linspace(0.01,0.99,200), 1,1))
ax.plot(np.linspace(0.01,0.99,200), st.beta.pdf(np.linspace(0.01,0.99,200), 5,2))
ax.plot(np.linspace(0.01,0.99,200), st.beta.pdf(np.linspace(0.01,0.99,200), 2,5))
ax.plot(np.linspace(0.01,0.99,200), st.beta.pdf(np.linspace(0.01,0.99,200), 5,5))
ax.set_title('Beta Distributions')
ax.legend(['Be(1/2,1/2)', 'Be(1,1)','Be(5,2)', 'Be(2,5)', 'Be(5,5)'])
```

```python
def gamma_pdf(x, lamb, sigma, nu):
    
    density = np.exp((nu - 1) * np.log(x - lamb) - (nu*np.log(sigma) + gammaln(nu)) - ((x - lamb) / sigma))
    
    return density
```

```python
# Gamma Distributions
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.plot(np.linspace(0,30,200), gamma_pdf(np.linspace(0,30,200), 0,1,2))
ax.plot(np.linspace(0,30,200), gamma_pdf(np.linspace(0,30,200), 0,10,1))
ax.plot(np.linspace(0,30,200), gamma_pdf(np.linspace(0,30,200), 0,5,2))
ax.plot(np.linspace(0,30,200), gamma_pdf(np.linspace(0,30,200), 0,2,5))
ax.plot(np.linspace(0,30,200), gamma_pdf(np.linspace(0,30,200), 0,4,4))
ax.set_title('Gamma Distributions')
ax.legend(['Ga(10,20)', 'Ga(10,1)','Ga(5,2)', 'Ga(2,5)', 'Ga(4,4)'])
```

```python
# Poisson Approximation of Binomial
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.bar(np.linspace(0,17,18), st.binom.pmf(np.linspace(0,17,18), n=30, p=0.2), width=0.05, color='k', zorder=1)
ax.scatter(np.linspace(0,17,18), st.poisson.pmf(np.linspace(0,17,18), 6), color='r', zorder=10)
```
