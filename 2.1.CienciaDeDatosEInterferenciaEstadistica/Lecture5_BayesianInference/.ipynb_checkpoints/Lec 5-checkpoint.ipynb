```python
# Note: This example is different from the one in the book, as I thought this one is more intuitive.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.special import gamma, gammaln, gammasgn
pd.options.display.max_rows=1000
```

```python
true_proportion = 0.40
sample_size = 100
n_tosses = 1 # Binomial (n,p) with n=1 is Bernoulli(p)
bern_draws = np.random.binomial(n_tosses, true_proportion, size=sample_size)
```

```python
np.sum(bern_draws) / sample_size
```

```python
# Prior distribution
fig, ax = plt.subplots(1, 1, figsize=(20,10))
plt.xticks(np.linspace(0,1,11))
mu = 0.5
M = 10
x_grid = np.linspace(0,1,101)
ax.plot(x_grid, st.beta.pdf(x_grid, M * mu, M * (1 - mu)), 'r-', lw=4, alpha=0.6, label='prior pdf')
```

```python
# Posterior distribution on p
success_count = 0
failure_count = 0
n_draws = 100
plt.rcParams["axes.prop_cycle"] = plt.cycler("color", plt.cm.RdBu(np.linspace(0,1,n_draws)))
fig, ax = plt.subplots(1, 1, figsize=(20,10))
plt.xticks(np.linspace(0,1,21))
mu = 0.5
M = 10
x_grid = np.linspace(0,1,101)
ax.plot(x_grid, st.beta.pdf(x_grid, M * mu, M * (1 - mu)), 'r-', lw=4, alpha=0.6, label='prior pdf')
for draw in bern_draws[:min(n_draws, sample_size)]:
    success_count += draw
    failure_count += 1 if draw == 0 else 0
    ax.plot(x_grid, st.beta.pdf(x_grid, success_count + M * mu, failure_count + M * (1 - mu)), '-', lw=3, alpha=0.2)

ax.plot(x_grid, st.beta.pdf(x_grid, success_count + M * mu, failure_count + M * (1 - mu)), 'b-', lw=4, alpha=0.6, label='posterior pdf')
```

```python
# Prior predictive distribution on k given M, mu
prob_k_prior = list()
for succ_count in range(0, sample_size + 1):
    fail_count = sample_size - succ_count
    prob = (gamma(sample_size + 1) / (gamma(succ_count + 1) * gamma(fail_count + 1))) \
            * (gamma(M) / (gamma(M*mu) * gamma(M*(1-mu)))) \
            * ((gamma(succ_count + M*mu) * gamma(fail_count + M*(1-mu))) / gamma(sample_size + M))                                              
    prob_k_prior.append(prob)
    
# Posterior predictive distribution on k given M, mu
prob_k_posterior = list()
for succ_count in range(0, sample_size + 1):
    fail_count = sample_size - succ_count
    prob = np.exp(gammaln(sample_size + 1) - gammaln(succ_count + 1) - gammaln(fail_count + 1) \
           + gammaln(sample_size + M) - gammaln(success_count + M*mu) - gammaln(failure_count + M*(1-mu)) \
           + gammaln(succ_count + success_count + M*mu) + gammaln(fail_count + failure_count + M*(1-mu)) - gammaln(sample_size + sample_size + M))                                          
    prob_k_posterior.append(prob)
```

```python
prior_pred = [(key, value) for key, value in enumerate(prob_k_prior)]
posterior_pred = [(key, value) for key, value in enumerate(prob_k_posterior)]
fig, ax = plt.subplots(1, 1, figsize=(20,10))
ax.scatter([i[0] for i in prior_pred], [i[1] for i in prior_pred])
ax.scatter([i[0] for i in prior_pred], [i[1] for i in posterior_pred])
```
