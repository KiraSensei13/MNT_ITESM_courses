```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.discrete.discrete_model import Logit
from statsmodels.tools.tools import add_constant
```

```python
spam_df = pd.read_csv('spam.csv')
```

```python
spam_df.head()
```

```python
logit_mod = Logit(spam_df['spam'], 
                  add_constant(spam_df.drop('spam',1))).fit()
logit_mod.summary()
```

```python
logit_mod = Logit(spam_df['spam'], 
                  add_constant(spam_df.drop(['spam', 'all', '3d', 'order', 'mail', 'email',
                                             'hpl', '1999', 'pm', 'conference', 'ch;'],1))).fit_regularized(alpha=15)
logit_mod.summary()
```
