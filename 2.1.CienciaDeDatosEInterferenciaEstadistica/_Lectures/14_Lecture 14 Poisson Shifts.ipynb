```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
x_grid = np.linspace(0,30,31)
mu_nought = 3

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.plot(st.poisson.pmf(x_grid, mu_nought),'.-')

mu_vec = np.array([6,9,12,15,18])
format_vec = ['.-y', '.-c', '.-g' ,'.-b' ,'.-r']
df = pd.DataFrame({'mu':mu_vec, 'format':format_vec})

for _, row in df.iterrows():
    alpha = np.log(row['mu']/mu_nought)
    psi = mu_nought * (np.exp(alpha) - 1)
    tilt_poi_pmf = st.poisson.pmf(x_grid, mu_nought) * np.exp(alpha * x_grid)
    tiltscale_poi_pmf = tilt_poi_pmf / np.exp(psi)
    ax.plot(tiltscale_poi_pmf, row['format'])
```
