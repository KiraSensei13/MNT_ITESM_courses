```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.discrete.discrete_model import Logit
from statsmodels.regression.linear_model import OLS
```

```python
# We start from  life tables A and B from our Lecture 21 output (check updated version of the lecture's Jupyter Notebook)
df = pd.read_csv('life_table_A.csv', usecols=['month_all', 'y', 'n', 'Arm'])
dfB = pd.read_csv('life_table_B.csv', usecols=['month_all', 'y', 'n', 'Arm'])
df = pd.concat([df, dfB])
df.rename(columns={'n':'At risk', 'y':'Died'}, inplace=True)
df = df[(df['month_all'] >= 7) & (df['month_all'] <= 12)]
```

```python
pt = pd.pivot_table(df, values=['At risk', 'Died'], index='month_all', columns='Arm')
pt = pt[[('At risk', 'B'), ('Died', 'B'), ('At risk', 'A'), ('Died', 'A')]]
pt
```

```python
# Let's compute the hypergeometric expectations
pt['E_deaths_A'] = np.round((pt[('At risk', 'A')] * (pt[('Died', 'A')] + pt[('Died', 'B')])) / pt.sum(axis=1),2)
```
