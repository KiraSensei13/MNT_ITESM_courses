```python
# import countries.csv
data = pd.read_csv("countries.csv");
print(data);
```

```python
# Compare the opulation growth in the US and China
US = data[data.country == "United States"];
CH = data[data.country == "China"];

plt.plot(US.year, US.population / 10**6); # display data in millions
plt.plot(CH.year, CH.population / 10**6);
plt.legend(["United States", "China"]);
plt.xlabel("year");
plt.ylabel("population");
plt.show();
```

```python
# plot the relative growth
plt.plot(US.year, US.population / US.population.iloc[0] * 100);
plt.plot(CH.year, CH.population / CH.population.iloc[0] * 100);
plt.legend(["United States", "China"]);
plt.xlabel("year");
plt.ylabel("population growth (first year = 100)");
plt.show();
```
