### Table 9.1

```python
insurance["h_hat"] = insurance.y / insurance.n
```

```python
def s_hat(h_hat):
    s = 1
    svals = []
    for hval in h_hat:
        s *= (1 - hval)
        svals.append(s)
    return svals
```

```python
insurance["S_hat"] = s_hat(insurance.h_hat)
```
