# 8.4 Regression Trees

### Setup

```python
import pandas as pd
from sklearn import tree
import graphviz
```

```python
spam = pd.read_csv("../data/SPAM.csv")
```

```python
spam.head()
```

### Fit decision tree

```python
X = spam.drop(["spam", "testid"], axis=1)
y = spam.spam
```

```python
model = tree.DecisionTreeClassifier(criterion="gini", max_leaf_nodes=7)
model.fit(X, y)
```

### Figure 8.7

```python
dot_data = tree.export_graphviz(model, feature_names=X.columns)  
graph = graphviz.Source(dot_data)  
graph
```
