# 8.1 Logistic Regression

### Setup

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

### Spam data

```python
spam = pd.read_csv("../data/SPAM.csv")
```

```python
spam.head()
```

### Fit the logistic model

```python
spam["spam"] = spam.spam.astype("int")
```

```python
X = spam.drop(["spam", "testid"], axis=1)
# Standardize the inputs
X = (X - X.mean()) / np.sqrt(X.var(ddof=1))
X = sm.add_constant(X)
```

```python
spam_model = sm.Logit(spam.spam, X)
spam_res = spam_model.fit()
```
