### Exercise 1 & 2 - part A
___

Let's calculate $$ \left(\frac{\partial V}{\partial T}\right)_P $$
from $$ dP=\left(\frac{\partial P}{\partial V}\right)_TdV+\left(\frac{\partial P}{\partial T}\right)_VdT $$
supposing that $$ P(V, T)=constant $$
so that $$ dP=0 $$
if $$ \left(\frac{\partial P}{\partial V}\right)_TdV+\left(\frac{\partial P}{\partial T}\right)_VdT=0 $$
then $$ \left(\frac{\partial V}{\partial T}\right)_P=-\frac{\left(\frac{\partial P}{\partial T}\right)_V}{\left(\frac{\partial P}{\partial V}\right)_T} $$

Let's use Mathematica to compute $$ \left(\frac{\partial P}{\partial V}\right)_T $$

```python
D[Solve[(P + a/V^2)*(V - b) == R*T, P], V] //ToRadicals //First //First //FullSimplify
```

Let's compute $$ \left(\frac{\partial P}{\partial T}\right)_V $$

```python
D[Solve[(P + a/V^2)*(V - b) == R*T, P], T] //ToRadicals //First //First //FullSimplify
```

Let's compute $$ \left(\frac{\partial V}{\partial T}\right)_P $$

```python
-(R/(-b + V))/(-((R T)/(b - V)^2) + (2 a)/V^3) //ToRadicals //FullSimplify
```

<div class="alert alert-block alert-info">
$$ \left(\frac{\partial V}{\partial T}\right)_P = \frac{R V^3 (V-b)}{-2 a (b-V)^2 R T V^3} $$
</div>

### Exercise 1 & 2 - part B
___

Let's calculate $$ \left(\frac{\partial V}{\partial P}\right)_T $$
from $$ dP=\left(\frac{\partial T}{\partial V}\right)_PdV+\left(\frac{\partial T}{\partial P}\right)_VdP $$
supposing that $$ T(V, P)=constant $$
so that $$ dT=0 $$
if $$ \left(\frac{\partial T}{\partial V}\right)_PdV+\left(\frac{\partial T}{\partial P}\right)_VdP=0 $$
then $$ \left(\frac{\partial V}{\partial P}\right)_T=-\frac{\left(\frac{\partial T}{\partial P}\right)_V}{\left(\frac{\partial T}{\partial V}\right)_P} $$

Let's use Mathematica to compute $$ \left(\frac{\partial T}{\partial V}\right)_P $$

```python
D[Solve[(P + a/V^2)*(V - b) == R*T, T], V] //ToRadicals //First //First //FullSimplify
```

Let's compute $$ \left(\frac{\partial T}{\partial P}\right)_V $$

```python
D[Solve[(P + a/V^2)*(V - b) == R*T, T], P] //ToRadicals //First //First //FullSimplify
```

Let's compute $$ \left(\frac{\partial V}{\partial P}\right)_T $$

```python
-((-b + V)/R)/((2 a b - a V + P V^3)/(R V^3)) //ToRadicals //FullSimplify
```

<div class="alert alert-block alert-info">
$$ \left(\frac{\partial V}{\partial P}\right)_T = \frac{(b-V) V^3}{2 a b - a V + P V^3} $$
</div>

### Exercise 3 - part A
___

The heat transferred from the hot to the cold aluminum piece $dq_{h \rightarrow c}$ is equal to the heat adsorbed by the cold from the hot $dq_{c \rightarrow h}$ after they reach the final temperature, $T_f$. $dq_{h \rightarrow c} = -dq_{c \rightarrow h}$ <br>
also,
if $m$ is the mass of aluminium, then $n = \frac{m}{M}$; where $M$ is the atomic weight of aluminium. <br>
$T_h = T_{A,0} = 400 K$ <br>
$T_c = T_{B,0} = 300 K$

Since $ dq = C_pdT = n\bar{C_p}dT $, the following can be done:
$$ C_{p_{cold}}dT = -C_{p_{hot}}dT $$
$$ n \int_{T_c}^{T_f} \overline{C_{p_{\text{Al}}}} \, dT+n \int_{T_h}^{T_f} \overline{C_{p_{\text{Al}}}} \, dT=0 $$
$$ n \left(T_f-T_c\right) \overline{C_{p_{\text{Al}}}}+n \left(T_f-T_h\right) \overline{C_{p_{\text{Al}}}} = 0 $$
$$ T_f = \frac{1}{2} \left(T_c+T_h\right) $$
$$ T_f = \frac{1}{2} \left(300 K+400 K\right) $$

<div class="alert alert-block alert-info">
$$ T_f = 350 K $$
</div
