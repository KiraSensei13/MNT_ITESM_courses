### Exercise 1 & 2 - part A

Let's calculate $$ \left(\frac{\partial V}{\partial T}\right)_P $$
from $$ dP=\left(\frac{\partial P}{\partial V}\right)_TdV+\left(\frac{\partial P}{\partial T}\right)_VdT $$
supposing that $$ P(V, T)=constant $$
so that $$ dP=0 $$
if $$ \left(\frac{\partial P}{\partial V}\right)_TdV+\left(\frac{\partial P}{\partial T}\right)_VdT=0 $$
then $$ \left(\frac{\partial V}{\partial T}\right)_P=-\frac{\left(\frac{\partial P}{\partial T}\right)_V}{\left(\frac{\partial P}{\partial V}\right)_T} $$

Let's use Mathematica to compute $$ \left(\frac{\partial P}{\partial V}\right)_T $$

```python
D[Solve[(P + a/V^2)*(V - b) == R*T, P], V] //ToRadicals //First //First //FullSimplify
```

Let's compute $$ \left(\frac{\partial P}{\partial T}\right)_V $$

```python
D[Solve[(P + a/V^2)*(V - b) == R*T, P], T] //ToRadicals //First //First //FullSimplify
```

Let's compute $$ \left(\frac{\partial V}{\partial T}\right)_P $$

```python
-(R/(-b + V))/(-((R T)/(b - V)^2) + (2 a)/V^3) //ToRadicals //FullSimplify
```

<div class="alert alert-block alert-info">
$$ \left(\frac{\partial V}{\partial T}\right)_P = \frac{R V^3 (V-b)}{-2 a (b-V)^2 + R T V^3} $$
</div>

The heat transferred from the hot to the cold aluminum piece $dq_{h \rightarrow c}$ is equal to the heat adsorbed by the cold from the hot $dq_{c \rightarrow h}$ after they reach the final temperature, $T_f$. $dq_{h \rightarrow c} = -dq_{c \rightarrow h}$ <br>
also,
if $m$ is the mass of aluminium, then $n = \frac{m}{M}$; where $M$ is the atomic weight of aluminium. <br>
$T_h = T_{A,0} = 400 K$ <br>
$T_c = T_{B,0} = 300 K$

Let's asume $C_p = 0.91 J g^{-1} K^{-1}$, $m = 1 Kg$ and $M = 26.982 g mol^{-1}$. <br>
Therefore, $n = 37.062 mol$

Since $ dq = C_pdT = n\bar{C_p}dT $, the following can be done:
$$ C_{p_{cold}}dT = -C_{p_{hot}}dT $$
$$ n \int_{T_c}^{T_f} \overline{C_{p_{\text{Al}}}} \, dT+n \int_{T_h}^{T_f} \overline{C_{p_{\text{Al}}}} \, dT=0 $$
$$ n \left(T_f-T_c\right) \overline{C_{p_{\text{Al}}}}+n \left(T_f-T_h\right) \overline{C_{p_{\text{Al}}}} = 0 $$
$$ T_f = \frac{1}{2} \left(T_c+T_h\right) $$
$$ T_f = \frac{1}{2} \left(300 K+400 K\right) $$

<div class="alert alert-block alert-info">
$$ T_f = 350 K $$
</div>

### Exercise 3 - part D
___
$$ \Delta S = n C_p ln\left(\frac{T_f}{T_i}\right) $$

where: $$ T_f = T_i - x \text{ for A; & }T_f = T_i + x \text{ for B;} $$

```python
# Function to compute the "next" pressure value for IRREVERSIBLE processes
def S_A_(n, Cp, Ti, x):
    Tf = Ti - x;
    return n*Cp*(np.log(Tf/Ti));

def S_B_(n, Cp, Ti, x):
    Tf = Ti + x;
    return n*Cp*(np.log(Tf/Ti));

# Draw the plot's workspace
scale = 6;
plt.subplots(figsize=(3*scale, 2*scale)) 

# define constants
n = 1000 #g or 37.062 mol
Cp = 0.91 #J / g K
Tmin = -300 #K
Tmax = 400 #K
Tequi = 350 #K
x = np.linspace(Tmin, Tmax, 1000);

# for B
Ti_B = 300 #K
S_B = S_B_(n, Cp, Ti_B, x)
plt.plot(x, S_B, '--', linewidth=3, label='Entropy(B)');

# for A
Ti_A = 400 #K
S_A = S_A_(n, Cp, Ti_A, x)
#print(S_A)
plt.plot(x, S_A, '--', linewidth=3, label='Entropy(A)');

# for A+B
plt.plot(x, (S_A+S_B), '-', linewidth=3, label='Entropy(System)');

plt.xlabel('x [K]', fontsize=24);
plt.ylabel('ΔS [J/g]', fontsize=24);
plt.title("Figure 1: ΔS vs. x", size=24)
plt.legend(prop={'size': 18})
plt.grid('on')
display(plt);
```

The heat transferred from the hot to the cold aluminum piece (orange line) is equal to the heat adsorbed by the cold from the hot (blue line) after they reach the final temperature of $350 K$.

<div class="alert alert-block alert-info">
The point where Entropy(A) meets Entropy(B) corresponds maximum entropy at the equilibrium temperature of the system.
</div>

### Problem 4 - For B

Final pressure <br>

$C_v = \frac{3}{2}R$ & $\gamma = \frac{5}{3}$
$$ P_{i_B} {V_{i_B}}^{\gamma} = P_{f_B} {V_{f_B}}^{\gamma} $$
$$ P_{f_B} = \frac{P_{i_B} {V_{i_B}}^{\gamma}}{{V_{f_B}}^{\gamma}} $$
$$ P_{f_B} = \frac{(1 atm) {(5 L)}^{\frac{5}{3}}}{{(2 L)}^{\frac{5}{3}}} $$
$$ P_{f_B} = 4.605 atm = 466602 Pa $$

Final Temperature
$$ T_{f_B} = \frac{P_{f_B} V_{f_B}}{n R} $$

$$ T_{f_B} = 224.4699 K $$

Internal energy
$$ \Delta U_B = Q_B - W_B $$
where $Q_B = 0$, and for adiabatic processes,
$$ W_B = \frac{P_{f_B}V_{f_B} - P_{i_B}V_{i_B}}{\gamma - 1} $$
$$ W_B = \frac{(4.605 atm)(2 L) - (1 atm)(5 L)}{\frac{5}{3} - 1} $$
$$ W_B = 6.315 atm L = 639.709 J $$

### Problem 4 - For A

```python
# Function to compute the "next" pressure value
def P_p_(Pi, Vi, Vf, gamma):
    return (Pi*Vi**gamma)/(Vf**gamma);

def P_(Tf, n, R, Vf):
    return (Tf*n*R)/(Vf);

# Draw the plot's workspace
scale = 6;
plt.subplots(figsize=(3*scale, 2*scale)) 

# For P_B
Pi = 1 #atm
Vi = 5 #L
Vf = 2 #L
gamma = 5/3
V = np.linspace(Vi,Vf,1000);
P_B = P_p_(Pi, Vi, V, gamma)
plt.plot(V, P_B, '-', linewidth=3, label='B');

# For P_A
Pi = 1 #atm
Vi = 5 #L
Vf = 8 #L
gamma = 1
V = np.linspace(Vi,Vf,1000);
P_A = P_p_(Pi, Vi, V, gamma)
plt.plot(V, P_B, '-', linewidth=3, label='A');

# For Temps
n = 0.5 #mol
R = 0.08206 #atm L / K mol

Tf = 224.4699 #K
Vi = 1.75 #L
Vf = 9 #L
V = np.linspace(Vi,Vf,1000);
P = P_(Tf, n, R, V)
plt.plot(V, P, '--', linewidth=3, label='224.4699 K');

Tf = 121.862 #K
Vi = 1.75 #L
Vf = 7 #L
V = np.linspace(Vi,Vf,1000);
P = P_(Tf, n, R, V)
plt.plot(V, P, '--', linewidth=3, label='121.862 K');

Tf = 897.8796 #K
Vi = 7 #L
Vf = 9 #L
V = np.linspace(Vi,Vf,1000);
P = P_(Tf, n, R, V)
plt.plot(V, P, '--', linewidth=3, label='897.8796 K');

plt.xlabel('V [L]', fontsize=24);
plt.ylabel('P [atm]', fontsize=24);
plt.title("Figure 2: P vs. V", size=24)
plt.legend(prop={'size': 18})
plt.grid('on')
display(plt);
```

For Isothermic Processes,
$$ ds = \frac{P}{T} dV = n R \frac{dV}{V} $$
$$ \Delta S = n R ln \left(\frac{V_f}{V_i}\right) $$
$$ \Delta S = (1 mol) (8.314 J K^{-1} mol^{-1}) ln \left(\frac{2.4 L}{0.3 L}\right) $$
<div class="alert alert-block alert-info">
$$ \Delta S = 17.289 J K^{-1} $$
</div>
    
Considering a van der Waals diatomic gas,
$$ \left(P + \frac{n^2 a}{V^2}\right)(V - n b) = n R T $$
$$ P = \frac{n R T}{V - n b} - \frac{n^2 a}{V^2} $$

Let's use the Maxwell relationship approach to determine the partial derivative of entropy.
$$ \left(\frac{\partial S}{\partial V}\right)_T = \left(\frac{\partial P}{\partial T}\right)_V  = \frac{n R}{V - n b}$$
$$ dS =  \frac{n R}{V - n b} dV $$
$$ \Delta S = \int_{V_i}^{V_f} \frac{n R}{V-b n} \, dV $$
$$ \Delta S = n R \left[ ln(-n b + V_f) - ln(-n b + V_i) \right] $$
$$ \Delta S = (1 mol)(8.314 J K^{-1} mol^{-1})\left[ ln(-(1 mol)(0.052 L mol_{-1}) + 2.4 L) - ln(-(1 mol)(0.052 L mol_{-1}) + (0.3 L)) \right] $$

<div class="alert alert-block alert-info">
$$ \Delta S = 18.69 J K^{-1} $$
</div>
