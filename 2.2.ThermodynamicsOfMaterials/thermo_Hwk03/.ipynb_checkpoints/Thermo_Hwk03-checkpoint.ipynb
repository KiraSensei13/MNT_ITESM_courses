Solar collector $\varepsilon$ is given by:
$$ \varepsilon = 0.75\, -1.75 \left(\frac{T}{300 K}-1\right) $$

The Carnot engine efficiency $\eta$ is given by:
$$ \eta = \left(\frac{T}{300 K}-1\right) $$

The system's efficiency $\epsilon$ is given by:
$$ \epsilon = \varepsilon \eta $$
$$ \epsilon = -\frac{1.75 T^{2}}{(300 K)^{2}} + \frac{4.25 T}{300 K} - 1.5 $$

Let's find $\frac{\partial \epsilon}{\partial T}$
$$ \frac{\partial \epsilon}{\partial T} = -\frac{2(1.75)}{300^{2}} T + \frac{4.25}{300}$$
$$ \frac{\partial \epsilon}{\partial T} = \frac{17}{1200} - \frac{7}{180000} T $$

Let's find T for $\frac{\partial \epsilon}{\partial T} = 0$
$$ \frac{17}{1200} - \frac{7}{180000} T = 0 $$
$$ T = 354.286 K $$

$$ \varepsilon = 0.75\, -1.75 \left(\frac{T}{300 K}-1\right); \text{with } T = 354.286 K $$
$$ \varepsilon = 0.433 $$

***
### Function to compute the "next" pressure value for REVERSIBLE processes
<div class="alert alert-block alert-info">
$$ P=\frac{n R T^{\gamma }}{V^{\gamma }} $$
</div>


***
### Function to compute the "next" pressure value for IRREVERSIBLE processes
$$ \text{dU}=-P \text{dV} =n \overline{C_v}\text{dT} $$
$$ n \overline{C_v} \int_{T_i}^{T_f} \, dT = -P_f \int_{V_i}^{V_f} \, dV$$
$$ n \overline{C_v} (T_f - T_i) = -P_f (V_f - V_i) $$
$$ n \overline{C_v} \left(\frac{P_f V_f}{n R} - \frac{P_i V_i}{n R}\right) = -P_f (V_f - V_i) $$
$$ \frac{n \overline{C_v} P_f V_f}{n R} - \frac{n \overline{C_v} P_i V_i}{n R} = -P_f (V_f - V_i) $$
$$ P_f \left( \frac{n \overline{C_v} V_f}{n R} + V_f - V_i \right) = \frac{n \overline{C_v} P_i V_i}{n R} $$

<div class="alert alert-block alert-info">
$$ P_f = \frac{\overline{C_v} P_i V_i}{\overline{C_v} V_f + V_f R - V_i R} $$
</div>


```python
# Function to compute the "next" pressure value for REVERSIBLE processes
def P_rev(n, R, T, V, gamma):
    return (n*R*(T**gamma))/(V**gamma);

# Function to compute the "next" pressure value for IRREVERSIBLE processes
def P_irrev(Cv, Pi, Vi, Vf, R):
    return (Cv*Pi*Vi)/(Cv*Vf+Vf*R-Vi*R);

# Draw the plot's workspace
n = 6;
plt.subplots(figsize=(3*n, 2*n)) 

# Initial State, Constants & Final Volume
Ti = 304.65; #K
Vi = 0.01; #10 L
Vf = 0.05; #50 L
V = np.linspace(Vi,Vf,1000);
n = 1; #mol
R = 8.314; #J/K mol)

# Reversible Isothemic Process
gamma = 1;
Vf = V;
Pi = P_rev(n, R, Ti, Vi, gamma);
Pf = P_rev(n, R, Ti, V, gamma);
plt.plot(Vf/Vi, Pf/Pi, '-', linewidth=3, label='Reversible Isothemic Process');

# Reversible Adiabatic Process (monoatomic)
Cv = (3/2)*R;
Cp = Cv + R;
gamma = Cp/Cv;
Vf = V;
Pi = P_rev(n, R, Ti, Vi, gamma);
Pf = P_rev(n, R, Ti, V, gamma);
plt.plot(Vf/Vi, Pf/Pi, '-', linewidth=3, label='Reversible Adiabatic Process (monoatomic)');

# Reversible Adiabatic Process (diatomic)
Cv = (5/2)*R;
Cp = Cv + R;
gamma = Cp/Cv;
Vf = V;
Pi = P_rev(n, R, Ti, Vi, gamma);
Pf = P_rev(n, R, Ti, V, gamma);
plt.plot(Vf/Vi, Pf/Pi, '-', linewidth=3, label='Reversible Adiabatic Process (diatomic)');

# Reversible Adiabatic Process (polyatomic)
Cv = 3*R;
Cp = Cv + R;
gamma = Cp/Cv;
Vf = V;
Pi = P_rev(n, R, Ti, Vi, gamma);
Pf = P_rev(n, R, Ti, V, gamma);
plt.plot(Vf/Vi, Pf/Pi, '-', linewidth=3, label='Reversible Adiabatic Process (polyatomic)');

# Irreversible Adiabatic Process (monoatomic)
Cv = (3/2)*R;
Cp = Cv + R;
gamma = Cp/Cv;
#Tf = 243.839; #K (from exercice 1.d)
#Vf = (n*R*Tf)/(50700); #L
#V = np.linspace(Vi,Vf,1000);
Vf = V;
Pi = P_rev(n, R, Ti, Vi, gamma);
print(Pi)
Pf = P_irrev(Cv, Pi, Vi, Vf, R);
plt.plot(Vf/Vi, Pf/Pi, '-', linewidth=3, label='Irreversible Adiabatic Process (monoatomic)');

# Irreversible Adiabatic Process (diatomic)
Cv = (5/2)*R;
Cp = Cv + R;
gamma = Cp/Cv;
Vf = V;
Pi = P_rev(n, R, Ti, Vi, gamma);
Pf = P_irrev(Cv, Pi, Vi, Vf, R);
plt.plot(Vf/Vi, Pf/Pi, '-', linewidth=3, label='Irreversible Adiabatic Process (diatomic)');

# Irreversible Adiabatic Process (polyatomic)
Cv = 3*R;
Cp = Cv + R;
gamma = Cp/Cv;
Vf = V;
Pi = P_rev(n, R, Ti, Vi, gamma);
Pf = P_irrev(Cv, Pi, Vi, Vf, R);
plt.plot(Vf/Vi, Pf/Pi, '-', linewidth=3, label='Irreversible Adiabatic Process (polyatomic)');

plt.rcParams.update({'font.size': 14})
plt.xlabel('Vf/Vi', fontsize=18);
plt.ylabel('Pf/Pi', fontsize=18);
plt.title("Figure 1: Pf/Pi vs. Vf/Vi", size=18)
plt.legend(prop={'size': 14})
display(plt);
```

Let's denote some variables: <br>
$ dQ_H $ = heat transfer out of the hot reservoir <br>
$ dQ_C $ = heat transfer into the cold reservoir <br>
$ dW $ = work output by the engine <br>
$ dT_H $ = change in temperature of the hot reservoir ($ dT_H < 0$ as it's cooling down) <br>
$ T_{Ho} $ = $ T_1 $ = initial temperature of the hot reservoir <br>
$ dT_C $ = change in temperature of the cold reservoir ($ dT_C > 0$ as it's warms up) <br>
$ T_{Co} $ = $ T_2 $ = initial temperature of the cold reservoir <br>
$ T $ = final temperature of both reservoirs <br>
$ \eta $ = 2nd law efficiency given

As both reservoirs have temperature-independent heat capacities $C_1$ and $C_2$, then it follows that:
$$ dQ_H = -C_1 T_H $$
$$ dQ_C = C_2 T_C $$

The engine efficiency can be defined as:
$$ \varepsilon = \frac{dW}{dQ_H} = 1 - \frac{dQ_C}{dQ_H} = 1 + \frac{dT_C}{dT_H} $$
With the following Carnot limiting efficiency:
$$ \varepsilon_c = 1 - \frac{T_C}{T_H} $$

The 2nd law efficiency $\eta$ is assumed to be constant.
$$ \eta = \frac{W_{net,out}}{Q_H} = \frac{Q_H - Q_C}{Q_H} = 1 - \frac{Q_C}{Q_H} $$

Rearranging terms, $T_C$ can be defined as a function of $T_H$:
$$ T_H \frac{dT_C}{dT_H} + \eta T_C = (\eta - 1) T_H $$

The complementary solution of the previous is $A T_H^{-\eta}$ where $A$ is a constant to taylor the initial conditions. On the other hand, by trying a particular solution of the equation that is proportional to $T_H$, the following is obtained $(\eta - 1) T_H / (\eta + 1)$. Adding together these complementary and particular solutions and fitting to the initial temperatures of the two reservoirs gives:
$$ \frac{T_C}{T_{\text{Ho}}}={\left(\frac{T_{\text{Ho}}}{T_H}\right)}^{\eta } \left\{\frac{T_{\text{Co}}}{T_{\text{Ho}}}+\frac{1 -\eta }{1 + \eta}\left(1-{\left(\frac{T_H}{T_{\text{Ho}}}\right)}^{\eta +1}\right)\right\} $$

Let's set $T_C = T_H = T$  to find the common final temperature of the two reservoirs:
$$ \frac{T}{T_{Ho}} = {\left(\frac{(T_{Co} + T_{Ho}) + \eta (T_{Co} - T_{Ho})}{2 T_{Ho}}\right)}^{\frac{1}{1+\eta}} $$

<div class="alert alert-block alert-info">
$$ T = T_{Ho} {\left(\frac{(T_{Co} + T_{Ho}) + \eta (T_{Co} - T_{Ho})}{2 T_{Ho}}\right)}^{\frac{1}{1+\eta}} $$
</div>

where: <br>
$ T_{\min }=\sqrt{T_{\text{Co}} T_{\text{Ho}}} $ is the geometric average of the initial temperatures <br>
$ T_{\max }=\frac{T_{Co} + T_{Ho}}{2} $ is their aithmetic average

The total work output by the engine is the difference in heat transfers for the two reservoirs, as follows:
$$ W = Q_H - Q_C = C_1(T_{Ho} - T) - C_2(T - T_{Co}) $$

<div class="alert alert-block alert-info">
And the maximum amount of work is given by:
$$ W_{max} = C_1 C_2 (T_{max} - T_{min}) $$
</div>

The maximum work is achieved for reversible operation of the engine when $\eta = 1$.  On the other hand, zero work is output for maximally irreversible operation of the engine when $\eta = 0$.
