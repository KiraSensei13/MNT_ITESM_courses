## 1.12 QUANTITATIVE EXAMPLE PROBLEMS

```python
def P_(n, R, T, V):
    numer = n*R*T;
    denom = V;
    return numer / denom;

scale = 6;
fig, ax = plt.subplots(figsize=(3*scale, 2*scale));

# constants
n = 1; #mole
R = 0.082057; #L atm/K mole
T = np.linspace(300, 1300, 1000) #K

# plot curves
V = 11.2; #L
P = P_(n, R, T, V);
plt.plot(T, P, '-', linewidth=3, label='11.2 liters')

V = 22.4; #L
P = P_(n, R, T, V);
plt.plot(T, P, '-', linewidth=3, label='22.4 liters')

V = 44.8; #L
P = P_(n, R, T, V);
plt.plot(T, P, '-', linewidth=3, label='44.8 liters')
    
# Plot config
plt.xscale('linear')
plt.yscale('linear')
plt.xlabel(r'$T$' + '    ' + r'$[K]$', fontsize=24, **pfont);
plt.ylabel(r'$P$' + '    ' + r'$[atm]$', fontsize=24, **pfont);
plt.title('', size=24, **pfont);
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.legend(prop={'size': 18});
display(plt);
```
