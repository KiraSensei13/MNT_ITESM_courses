* from [1], equation (1) "is the G-H equation in one of its most subtantial form",
$$ \left(\frac{\partial(G / T)}{\partial T}\right)_{P,n} = -\frac{H}{T^2} $$

* where <br>
$T$ = temperature <br>
$P$ = pressure <br>
$n$ = composition vector that contains the amoun of all the components of the thermodynamic system <br>
$G$ = Gibbs function <br>
$H$ = enthalpy <br>

$$ d \left(\frac{\partial(G / T)}{\partial T}\right)_{P,n} = -\frac{H}{T^2} dT $$
$$ \int_{T_i}^{T_f} \, d\left(\frac{\partial(G / T)}{\partial T}\right)_{P,n} = \int_{T_i}^{T_f} -\frac{H}{T^2}\, dT $$
$$ \int_{T_i}^{T_f} \, d\left(\frac{\partial(G / T)}{\partial T}\right)_{P,n} = -H \int_{T_i}^{T_f} \frac{1}{T^2}\, dT $$

* if $\partial(G / T) = \Delta G_T$

$$ \int_{T_i}^{T_f} \, d\left(\frac{\Delta G_T}{T}\right)_{P,n} = -H \int_{T_i}^{T_f} \frac{1}{T^2}\, dT $$
$$ \left(\frac{\Delta G_{T_f}}{T_f}\right)_{P,n} - \left(\frac{\Delta G_{T_i}}{T_i}\right)_{P,n} = -H \left[\frac{1}{T_i} - \frac{1}{T_f}\right] $$

* Solve for $\Delta G_{T_f}$
$$ \left(\frac{\Delta G_{T_f}}{T_f}\right)_{P,n} = + \left(\frac{\Delta G_{T_i}}{T_i}\right)_{P,n} - H \left[\frac{1}{T_i} - \frac{1}{T_f}\right] $$
$$ \Delta G_{T_f} = + \frac{\Delta G_{T_i} T_f}{T_i} - H \left[\frac{T_f}{T_i} - \frac{T_f}{T_f}\right] $$
$$ \Delta G_{T_f} = \Delta G_{T_i} \frac{T_f}{T_i} - H \left[\frac{T_f}{T_i} - 1\right] $$

* therefore
<div class="alert alert-block alert-info">
$$ \Delta G_{T_f} = \Delta G_{T_i} \frac{T_f}{T_i} + H \left[1 - \frac{T_f}{T_i}\right] $$
</div>

$$ \Delta G = \Delta H - T \Delta S $$

$$ T = \frac{\Delta H - \Delta G}{S} $$

$$ T = \frac{90.46-(-58.43)}{\frac{205.2}{2} + 71.13 - 75.9} K $$


$$ T = 1.5219 K $$

* let's obtain the eqilibrium temperature (spontaneous reaction)

$$ \Delta G = \Delta H - T \Delta S = 0 $$

$$ \Delta H - T \Delta S = 0 \rightarrow \Delta H = T \Delta S $$

$$ \frac{\Delta H}{\Delta S} = T $$

* calculate $\Delta S$

$$ \Delta S = 75.9 + \frac{205.2}{2} - 71.13 $$

$$ \Delta S = 107.37 kJ mol^{-1} K^{-1} $$

* calculate $\Delta H$

$$ \Delta H = 0+\frac{1}{2} (0) - (-90.46) $$

$$ \Delta H = 90.46 J mol^{-1} $$
         
* substitute $\Delta S$ and $\Delta H$

$$ T = \frac{\Delta H}{\Delta S} $$

$$ T = \frac{90.46}{107.37 \times 10^{-3}} $$


$$ T = 842.82 K $$


* therefore
<div class="alert alert-block alert-info">
The reaction will be spontaneous at $T$ less than $842.82 K$
</div>
