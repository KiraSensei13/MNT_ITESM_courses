<h1>Binary Phase Diagrams</h1>

````python
# PYTHON LIBRARIES
%matplotlib inline

import numpy as np
np.seterr(divide='ignore', invalid='ignore')
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('xtick', labelsize=15)     
plt.rc('ytick', labelsize=15)

from scipy.optimize import least_squares
````

$$ \Delta G_{m,A} = 8000 - 10 T [J/mol] $$
$$ \Delta G_{m,B} = 12000 - 10 T [J/mol] $$

````python
def DG_mA_(T):
    return 8000 - 10*T;

def DG_mB_(T):
    return 12000 - 10*T;
````

$$ \Omega_l = -20000 [J/mol] $$
$$ \Omega_s = 0 [J/mol] $$

````python
Omega_l = -20000
Omega_s = 0
````

$$ \Delta G_s = - x_A \Delta G_{m,A} + R T (x_A ln x_A + x_B ln x_B) + \Omega_s x_A x_B  $$
$$ \Delta G_l = x_B \Delta G_{m,B} + R T (x_A ln x_A + x_B ln x_B) + \Omega_l x_A x_B $$

````python
def DG_s_(xb, T, Omega_s):
    DG_mA = DG_mA_(T)
    R = 8.3144
    xa = 1 - xb
    
    return -xa*DG_mA + R*T*(xa*np.log(xa) + xb*np.log(xb)) + Omega_s*xa*xb;

def DG_l_(xb, T, Omega_s):
    DG_mB = DG_mB_(T)
    R = 8.3144
    xa = 1 - xb
    
    return xb*DG_mB + R*T*(xa*np.log(xa) + xb*np.log(xb)) + Omega_l*xa*xb;
````

$$ G_{B,s} = 0 $$
$$ G_{A,l} = 0 $$

Substituing numerical values and deriving ...

$$ \Delta G_s = (10 T-8000) \left(1-x_B\right)+8.3144 T \left(\left(1-x_B\right) \log \left(1-x_B\right)+x_B \log \left(x_B\right)\right) $$
$$ \frac{d \Delta G_s}{d x_B} =  -8.3144 T \log \left(1-x_B\right)+8.3144 T \log \left(x_B\right)-10. T+8000. $$

$$ \Delta G_l = (12000-10 T) \left(1-x_B\right)+8.3144 T \left(\left(1-x_B\right) \log \left(1-x_B\right)+x_B \log \left(x_B\right)\right)-20000 x_B \left(1-x_B\right) $$
$$ \frac{d \Delta G_l}{d x_B} = -8.3144 T \log \left(1.\, -1. x_B\right)+8.3144 T \log \left(x_B\right)+40000. x_B+10. T-32000. $$

````python
def d_DG_s_(xb, T):
    return - 8.3144*T*np.log(1 - xb) + 8.3144*T*np.log(xb) - 10*T + 8000

def d_DG_l_(xb, T):
    return - 8.3144*T*np.log(1 - xb) + 8.3144*T*np.log(xb) + 40000*xb + 10*T - 32000
````

Find common tangents ... <br>

```python
DG_s_(xb, T, Omega_s) and DG_l_(xb, T, Omega_s)
```
have a common tangent line at $xb = a$ if ... <br>
```python
DG_s_(a, T, Omega_s) = DG_l_(a, T, Omega_s)
```
and
```python
d_DG_s_(a, T) = d_DG_l_(a, T)
```
---
Let's solve $a$ from the 1st condition:
```python
d_DG_s_(a, T) = 0
```


````python
T = 600
xb = np.linspace(0.0, 1.0, 1000) 

# PLOT FIG
scale = 6;
fig, ax = plt.subplots(figsize=(3*scale, 2*scale));

# Plot
#plt.scatter(T, C, s=25, color='red', label='Raw data');
x = xb
y = DG_l_(xb, T, Omega_l)/1000
plt.plot(x, y, '-', linewidth=3, label='liquid')

x = xb
y = DG_s_(xb, T, Omega_s)/1000
plt.plot(x, y, '-', linewidth=3, label='solid')

#x = xb
#y = d_DG_d_T_(xb)
#plt.plot(x, y, '--', linewidth=3, label='solid')

#ax.set(autoscale_on=False)
#plt.plot(T, C_prev, '--', linewidth=1, label='pre-Fitted curve')

# Print fitting parameters as plot legends
x = x[0]
y = y[0]
plt.scatter(x, y, s=0, label=r'$T = $' + str(round(T, 2)) + r'$K$')

# Display plots
plt.yscale('linear');
plt.xlabel(r'$x_B$', fontsize=24);
plt.ylabel(r'$G$' + '    ' + r'$[kJ \cdot mol^{-1}]$', fontsize=24);
plt.title('Figure 1', size=24);
plt.legend(prop={'size': 18});
display(plt);
````
