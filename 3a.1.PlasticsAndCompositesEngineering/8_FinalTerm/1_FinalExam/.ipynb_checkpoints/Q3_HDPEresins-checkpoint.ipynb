### Get the activation energy $E_H$ and $E_V$
### $$ \log a_T \textit{ & } \log b_T \textit{ vs. }  \left( \frac{1}{T} - \frac{1}{T_0} \right) $$
### Plot $\log a_T$ and $\log b_T$ vs. $\left( \frac{1}{T} - \frac{1}{T_0} \right)$ and fit a linear regression to get $E_H$ and $E_V$

```python
# Declare sample temperatures in Celsius and Kelvin
T_C    = pd.Series([170, 190, 210, 230]);
T_K    = T_C + 273.15;
T_Cref = T_C[2];
T_Kref = T_K[2];

E_H, E_V = jbr.calculate_activationEnergies(T_K, T_Kref, aT, bT)
```

---
### Compute the shift factor $a_T$ for arbitrary temperatures with the Williams-Landell-Ferry (WLF) equation
### $$ a_T = \exp \left[ E_H \left( \frac{1}{T} - \frac{1}{T_0} \right) \right] $$
### $$ b_T = \exp \left[ E_V \left( \frac{1}{T} - \frac{1}{T_0} \right) \right] $$

```python
newTemperature = 200
aT_temp = jbr.calculate_shiftFactorWithWLF(E_H, newTemperature + 273.15, T_Kref)
bT_temp = jbr.calculate_shiftFactorWithWLF(E_V, newTemperature + 273.15, T_Kref)

aT = aT.append(pd.Series([aT_temp]));
bT = bT.append(pd.Series([bT_temp]));
T_C = T_C.append(pd.Series([newTemperature]));
T_K = T_C + 273.15;

display('a_T'); display(aT);
display('b_T'); display(bT);
display('T_C'); display(T_C);
```

```python
jbr.plot(
    dataframe = df,
    pltname   = "Loss Modulus vs frequency (" + r'$T_{ref} = 210°C$' + ")",
    x_str     = ["$a_T \omega_{170°C}$", "$a_T \omega_{190°C}$", "$a_T \omega_{210°C}$", "$a_T \omega_{230°C}$"],
    x_units   = r"$\left[ \dfrac{rad}{s} \right]$",
    y_str     = ["$b_T {G^{\prime\prime}(\omega)}_{170°C}$", "$b_T {G^{\prime\prime}(\omega)}_{190°C}$", "$b_T {G^{\prime\prime}(\omega)}_{210°C}$", "$b_T {G^{\prime\prime}(\omega)}_{230°C}$"],
    y_units   = r"$\left[ Pa \right]$")
```

### Compute fitting curve with the Wagner model with the master curve at $210°C$
### Shear viscosity as a function of a given shear rate $\dot{\gamma}_o$
### $$ \eta (t, \dot{\gamma}_o) = f_1 \sum_{i=1}^{n} \frac{a_i}{{\alpha_i}^2} \left( 1 - e^{1 - \alpha_i t} (1 - \dot{\gamma}_o n_1 \lambda_i \alpha_i t) \right) + f_2 \sum_{i=1}^{n} \frac{a_i}{{\beta_i}^2}  \left( 1 - e^{1 - \beta_i  t} (1 - \dot{\gamma}_o n_2 \lambda_i \beta_i  t) \right) $$
 
#### where:
#### $$ f_2 = 1 - f_1 $$
#### $$ \alpha_i = \frac{1 + n_1 \lambda_i \dot{\gamma}_o}{\lambda_i} $$
#### $$ \beta_i  = \frac{1 + n_2 \lambda_i \dot{\gamma}_o}{\lambda_i} $$
$\eta (t, \dot{\gamma}_o) =$ shear viscosity <br>
$t =$ time <br>
$\dot{\gamma}_o =$ shear rate <br>
$a_i =$ the $i^{th}$ elastic value of the Maxwell element <br>
$\lambda_i =$ the $i^{th}$ characteristic time of the Maxwell element <br>
$f_1, f_2, n_1, n_2$ are fitting parameters

### The Wagner model gives the shear viscosity vs. time of a single shear rate ...
### The steady state shear viscosity at several shear rates at $t = \infty$ is given by:
### $$ \eta (t = \infty, \dot{\gamma}_o) = f_1 \sum_{i=1}^{n} \frac{a_i}{{\alpha_i}^2} + f_2 \sum_{i=1}^{n} \frac{a_i}{{\beta_i}^2} $$

```python
parameters = jbr.plot_Wagner_fit_viscosity(
    df_master_eta, 
    "$_{master} \omega_{210°C}$",
    "$_{master} {\eta^{*}(\omega)}_{210°C}$")
```

### Steady state first normal stress difference $N1$ at $t = \infty$
### $$ N1(t, \dot{\gamma}_o) = {\dot{\gamma}_o}^2 \left\{ f_1 \sum_{i=1}^{n} a_i {\alpha_i}^3 \left( 1 - e^{- \alpha_i t} \left[ 1 + \alpha_i t - {\alpha_i}^2 \left( n_1 \lambda_i \frac{\dot{\gamma}_o}{2} \right) t^2 \right) \right] + f_2 \sum_{i=1}^{n} a_i {\beta_i}^3 \left( 1 - e^{- \beta_i t} \left[ 1 + \beta_i t - {\beta_i}^2 \left( n_2 \lambda_i \frac{\dot{\gamma}_o}{2} \right) t^2 \right) \right] \right\} $$
### $$ N1(t=\infty, \dot{\gamma}_o) = {\dot{\gamma}_o}^2 \left\{ f_1 \sum_{i=1}^{n} a_i {\alpha_i}^3 + f_2 \sum_{i=1}^{n} a_i {\beta_i}^3 \right\} $$
