#### 1. Calculate the total surface when you divide a cube with an 1 cm edge:

• 8 smaller cubes. <br>
• 16 smaller cubes. <br>
• 32 smaller cubes. <br>
• 64 smaller cubes. <br>
• 128 smaller cubes. <br>
• 256 smaller cubes. <br>
• 512 smaller cubes.

```python
def cubeSurfArea(length):
    return 6*(length**2); #cm^2

def subCubeLength(volume, noSubCubes):
    subCubeVolume = volume/noSubCubes;
    return (subCubeVolume)**(1/3); #cm

def totalSurfArea(volume, noSubCubes):
    length = subCubeLength(volume, noSubCubes);
    return noSubCubes*cubeSurfArea(length); #cm^2
```

```python
volume         = 1 #cm
noSubCubes_dp  = np.array([8, 16, 32, 64, 128, 256, 512]);
surfaceArea_dp = totalSurfArea(volume, noSubCubes_dp);
noSubCubes     = np.linspace(1, 512, 100);
surfaceArea    = totalSurfArea(volume, noSubCubes);

scale = 6;
plt.subplots(figsize=(3*scale, 2*scale));
plt.scatter(noSubCubes_dp, surfaceArea_dp);
plt.plot(noSubCubes, surfaceArea, linewidth=2);
plt.xlabel("Number of cubes", fontsize=24);
plt.ylabel("Total surface area", fontsize=24);
```

#### 1. How many iron atoms can be inside a cube with a 30 nm, 10 and 3nm edge?

```python
ironAtomicRadius = 0.14; #nm
ironAtomicVolume = (4/3)*np.pi*(ironAtomicRadius**3); #nm^3

edgelength_dp       = np.array([30, 10, 3]); #nm
cubeVolume_dp       = edgelength_dp**3; #nm^3
noOfAtomsInACube_dp = cubeVolume_dp/ironAtomicVolume; #atoms
edgelength          = np.linspace(30, 3, 100); #nm
cubeVolume          = edgelength**3; #nm^3
noOfAtomsInACube    = cubeVolume/ironAtomicVolume; #atoms

scale = 6;
plt.subplots(figsize=(3*scale, 2*scale));
plt.scatter(edgelength_dp, noOfAtomsInACube_dp);
plt.plot(edgelength, noOfAtomsInACube, linewidth=2);
plt.xlabel("Cube edge length", fontsize=24);
plt.ylabel("No of iron atoms within the cube", fontsize=24);
```

#### 2. How many iron atoms are on the surface of each cube?

```python
ironAtomicArea = np.pi*(ironAtomicRadius**2); #nm^2

cubeSurfAreas_dp    = cubeSurfArea(edgelength_dp); #nm^2
noOfAtomsInAEdge_dp = cubeSurfAreas_dp/ironAtomicArea; #atoms

scale = 6;
plt.subplots(figsize=(3*scale, 2*scale));
plt.scatter(edgelength_dp, noOfAtomsInAEdge_dp);
# plt.plot(edgelength, noOfAtomsInACube, linewidth=2);
plt.xlabel("Cube edge length", fontsize=24);
plt.ylabel("No of iron atoms within the cube's surface", fontsize=24);
```
